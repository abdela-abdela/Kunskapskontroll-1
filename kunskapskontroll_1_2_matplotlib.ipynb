```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

- It could either be done in object-oriented approach or in pyplot approach. The object-oriented approach is most recommended for it's flexibility and handling of most complicated visualization and can be reused in bigger projects.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

- A figure is the overall window that keeps track of everything that a chart is made of.
- An axes is a subplot in a figure where a chart is drawn.
- An axis is the individual x-axis or y-axis in a figure.
- An artist is everything visual in a figure. From the lines, images, titles, labels etc.

### When plotting in Matplotlib, what is the expected input data type?

- When plotting, the most expected and frequently used data type is numpy arrays.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 30)
y = x**2

fig, ax = plt.subplots() # Explicita metoden.
ax.plot(x, y, label='y = x^2')
ax.set_title('Explicit metod: Plot av funktionen y = x^2')
ax.set_xlabel('x-axeln')
ax.set_ylabel('y-axeln')

plt.figure() # Implicita metoden.
plt.plot(x, y, label='y = x^2')
plt.title('Implicit metod: Plot av funktionen y = x^2')
plt.xlabel('x-axeln')
plt.ylabel('y-axeln')

ax.plot(x, y)
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (plot1, plot2) = plt.subplots(1, 2, figsize=(12, 6))

plot1.scatter(random_data_x, random_data_y, alpha=0.5)
plot1.set_title('Scatter Plot')
plot1.set_xlabel('Data X')
plot1.set_ylabel('Data Y')

plot2.bar(names, values, color=['purple', 'red', 'orange', 'yellow', 'green'])
plot2.set_title('Bar plot')
plot2.set_xlabel('Fruits')
plot2.set_ylabel('Quantity')

plt.show()
```
